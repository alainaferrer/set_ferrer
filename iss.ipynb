```python
import requests
import matplotlib.pyplot as plt
import time
```

```python
# Track Location of ISS
given_api= 'https://api.wheretheiss.at/v1/satellites/25544'

# Empty lists to store data points
timestamps = []
latitudes = []
longitudes = []

# Get the Location Data

for i in range(1,101): #Fetch data every 10 seconds
    response = requests.get(given_api)
    data = response.json()
    timestamps.append(data['timestamp'])
    latitudes.append(data['latitude'])
    longitudes.append(data['longitude'])
    print(data['timestamp'], data['latitude'], data['longitude'])
    time.sleep(10)

# Plotting the ISS path
plt.figure(figsize=(10, 6))
plt.plot(longitudes, latitudes, marker='o', linestyle='-', color='b')
plt.title('International Space Station Path')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.grid(True)
plt.tight_layout()
plt.gca().invert_xaxis()  # invert x-axis to show correct map orientation
plt.show()
```
